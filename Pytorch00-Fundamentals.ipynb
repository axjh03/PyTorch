**TRANSPOSE**

---
*Tackling shape erros*


````python
# Creating two matrix and then doing multiplication
# (3,2) * (3,2)
# Since the inner dimensions are not matching the matrix will not be multiplied and will throw an error
````

````python
tensor_A = torch.randint(high=10, size=(3,2), dtype=torch.int8)
tensor_B = torch.randint(high=10, size=(3,2), dtype=torch.int8)
try:
  tensor_A @ tensor_B
except:
  print("Size didn't match up")

# Now trying to transpose the matrix
# so (3,2) --> (2,3) then only it will work

tensor_B = torch.transpose(tensor_B, 0, 1)
# OR
# tensor_B = tensor_B.T
"""
dim0: Corresponds to the rows (axis 0).
dim1: Corresponds to the columns (axis 1).
"""

print(tensor_B.shape)
try:
  tensor_A @ tensor_B
  print("Worked")
except:
  print("Size didn't match up")



# We can either use
# @ sign for matrix multiplication or
# torch.matmul()
# or torch.mm()
````

## **Reshaping, stacking, squeezing and unsqueezing tensors**

#### * Reshape - Reshapes an input tensor to defined shaped
#### * View - View a tensor in different shape but don't change anything in memeory / the original tensor
#### * Stacking - Combine multiple tensor on top of each other ( h-stack ) or side by side ( v - stack ) or just torch.stack
#### * Squeeze - Removes all 1 dimensions from a tensor
#### * Unsqueeze - Adds all 1 dimensions to target tensors
#### * Permute - Return a view of input with dimensions permutted ( swapped ) in a different way

````python
# Let's create a tensor
import torch
x = torch.arange(1,10)
x,x.shape
````

# 1*9 = 9
# We have 9 elements and if we multiply both we can't reshape more than that


````python
# Add an Extra dimension
x_reshaped = x.reshape(1,9) # We add an extra 1 dimension

# 1*9 = 9
# We have 9 elements and if we multiply both we can't reshape more than that

# so 3 * 3 = 9 and hence will work
x_reshaped2 = x.reshape(3,3) # Should work
"Reshaped to 1,9","",x_reshaped, x_reshaped.shape,"","Reshaped to 3,3", x_reshaped2, x_reshaped2.shape
````

````python
# ( 1, 9) is (row, column)
# What if we reshape that to (9,1) it will have 9 row and 1 column
x_reshaped3 = x.reshape(9,1)
x_reshaped3, x_reshaped3.shape
````

````python
# Change the view
z = x.view(1,9)
print(f"{z} \n {z.shape}")

# Z shares the same memory of x, it's just a different view and we are not assigning x to z as new variable

# Let's change the z
# z[:,0] will select all row's 0th column
# so we selected 1 here

z[:,0] = 100
print(f"x : {x}\nz : {z}")

# Changing z changes x because it's just a view
````

````python
# Change the view
z = x.view(3,3)
print(f"{z} \n {z.shape}")

z[0,0] = 99
print(f"x : {x}\nz : {z}")

# Changing z changes x because it's just a view
````

````python
# Stacking
x = torch.arange(3,10)
x_stacked = torch.stack([x,x,x,x], dim=0)
# 0 means stacking vertically i.e v stack
# 1 means stacking horizontally i.e h stack
print(x_stacked)
````

````python
# Squeeze & Unsqueeze
x = torch.ones(size=(1,4))
x,x.shape
````

````python
x = torch.squeeze(x, dim=0) # Row
x, x.shape
# Squeeze removes all single dimensions from the target tensor
````

````python
# Unsqueeze
print(x.shape)
x = torch.unsqueeze(x,dim=1) # Column
print(x.shape)
print(x)
````

````python
# Permute
# It's a VIEW and will share the same memory if we create another variable
x = torch.randn(2,3,5) # Two 3x5 random number
print(x.shape)
# Now we are gonna permute i.e swap the dimensions of the tensor
# 0 is first, 1 is second, 2 is third and we write it the way we want to rearrange
x=torch.permute(x,(2,1,0)) # this will change that to five 3x3 matrix
print(x.shape)
````

````python
# Importing stuffs
import numpy as np
import torch

# Creating a numPy array
array = np.arange(start=1, stop=10, step = 1)
print(type(array))

# Convert NumPy -> Torch
tensor = torch.from_numpy(array)
print(type(tensor))
# These doeesn't share memory
# Check dataType, doing back and forth might change the dataType

array, tensor
````

# Reproducibility
### Trying to take the random out of random
### In short how a neural networks works
` start with random numbers -> tensor operations -> update random numbers to try and make them better representation of the data -> again -> again .... `

````python
torch.rand(3,3) # generates a random
````

To reduce the randomness in neural networks and pytorch use will use *** random seed ***.



````python
import torch

# Create two random tensors

r_tensor_A = torch.rand(3,4)
r_tensor_B = torch.rand(3,4)
print(r_tensor_A == r_tensor_B)
r_tensor_A, r_tensor_B

# Every times random number is generated and almost impossible to get the same set of random again
````

````python
# Let's cerate some random but reproducible tensors
import torch

# Set a seed variable # We can set anything, it's like create a flavour of randomness
seed_value = 100
# Set the random seed
torch.manual_seed(seed_value) # Flavour no 100
# Every random created between two 'torch.manual_seed' these command will repeat itself if used same seed flavor i.e 100 in this case
random_tensor_C = torch.rand(3,4)

torch.manual_seed(seed_value)
random_tensor_D = torch.rand(3,4) # should be same as C
random_tensor_C,random_tensor_D, random_tensor_D==random_tensor_C
````

## IF A TENSOR is on GPU, we CAN'T TRANSFORM to NUMPY

````python
tensor_on_gpu.numpy()
````

#### So first we need to move tensor to CPU then convert then back to GPU

````python
import torch

# Assuming tensor_on_gpu is a PyTorch tensor on GPU

# Moving from GPU to CPU
tensor_back_on_cpu = tensor_on_gpu.to("cpu")
print(type(tensor_back_on_cpu))
print(tensor_back_on_cpu.device)

# Converting to NumPy
numpy_array = tensor_back_on_cpu.numpy()
print(type(numpy_array))

# Moving back to GPU
numpy_on_gpu = torch.from_numpy(numpy_array).to("cuda")
print(type(numpy_on_gpu))
print(numpy_on_gpu.device)
````
